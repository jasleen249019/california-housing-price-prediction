```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

```

```python

import pandas as pd

df = pd.read_excel("SALES.xlsx")
print(df.head())
```

```python

print("Available columns:", df.columns)


median_sales = df['SALES'].median()
mode_sales = df['SALES'].mode()[0]
std_sales = df['SALES'].std()

print("Mean_Sales:", mean_sales)
print("Median_Sales:", median_sales)
print("Mode of Sales:", mode_sales)
print("Standard Deviation of Sales:", std_sales)
```

```python
#FIT MULTIPLE REGRESSION MODEL USING OLS

X = df[['TV ADVERTISING', 'RADIO', 'NEWSPAPER']]
X = sm.add_constant(X)  # adds intercept term
y = df['SALES']

model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
#REGRESSION EQUATION

intercept = model.params['const']
b1 = model.params['TV ADVERTISING']
b2 = model.params['RADIO']
b3 = model.params['NEWSPAPER']

print(f"Regression Equation: Sales = {intercept:.4f} + {b1:.4f}*TV + {b2:.4f}*Radio + {b3:.4f}*Newspaper")
```

```python
#INTERPRET R SQUARE

print("R-square:", model.rsquared)
```

```python
#MODEL SIGNIFICANCE F-STATISTIC AND P-VALUE

print("F-statistic:", model.fvalue)
print("Model p-value:", model.f_pvalue)
```

```python
#SIGNIFICANCE

print(model.pvalues)
```

```python
from statsmodels.stats.stattools import durbin_watson

dw = durbin_watson(model.resid)
print("Durbin-Watson value:", dw)
```
